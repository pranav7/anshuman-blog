# Quick PDF, CDF Implementation in code
> For when math can look scary!

### Just some quick code to understand PDF and CDF from a code/viz perspective, for when math equations can look scary

```python
# Import all the nice packages
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import scipy.stats
```

```python
# Decide on a mean and sigma for the gaussian dist.
mean = 0
sigma = 2.5
# Generate random data to plot the gaussian
x = np.arange(-10,10,0.1)
```

```python
y_values = scipy.stats.norm(mean, sigma)
y_values_pdf = y_values.pdf(x)
y_values_cdf = y_values.cdf(x)
# Nice function to calculate integral
area = integrate.quad(lambda x: y_values.pdf(x), -10, 0)
area[0]
area_str = str(round(area[0], 2))
```

```python
# Matplotlib magic
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
plt.plot(x, y_values_pdf)
plt.text(-10, 0.1, "Area in highlighted region = probability of getting 0 or less = {}".format(area_str))
plt.fill_between(x, y_values_pdf, where = (x<0), color = "g")
plt.title("Probability Density Function")
plt.subplot(1,2,2)
plt.plot(x, y_values_cdf)
plt.axhline(y=0.5)
plt.axvline(x=0)
plt.title("Cumulative Density Function")
plt.show()
```

### Area under PDF is the probability. Any point on the CDF is the the probability of getting that value OR less.
### Example - The shaded green region is the probability of getting 0 or less from PDF ---> 0.5 in this case (half of fig is shaded)
### If you draw a vertical line at 0, it intersects the CDF at 0.5 ----> shows probability of getting 0 OR less is 0.5
